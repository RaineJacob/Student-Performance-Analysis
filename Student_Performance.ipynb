**In This Exploratory Data Analysis we are finding out the factors that affect students's Exam score So we can take further action according to it**.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

**Importing the dataset **

```python
data = pd.read_csv('/content/drive/MyDrive/Datasets/StudentPerformanceFactors.csv')
data.head(5)
```

```python
data.describe()
```

Removing Unwanted Columns from Dataset

```python
data = data.drop(columns=['Distance_from_Home', 'Peer_Influence', 'Teacher_Quality', 'Sleep_Hours'])
```

Demographic breakdown of Dataset

```python
gender_counts = data['Gender'].value_counts() #Counts the occurence of each genders

#creating pie chart
plt.figure(figsize=(5,5))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Gender')
plt.axis('equal')
plt.show()
```

**Distribution of Score **


As visible Majority fo score is around 70s

```python
sns.histplot(data['Exam_Score'], kde=True, bins=20)
plt.title('Distribution of Exam Scores')
plt.xlabel('Exam Score')
plt.ylabel('Count')
plt.show()
```

Performance By Gender

As graphs shows us Females are slightly better performing in Exams

```python
sns.boxplot(x='Gender', y='Exam_Score', data=data)
plt.title('Performance by Gender')
plt.xlabel('Gender')
plt.ylabel('Exam Score')
plt.show()
```

# Effect of Study hours on Score

```python
sns.scatterplot(x='Hours_Studied', y='Exam_Score', hue='Gender', data=data)
plt.title('Hours Studied vs Exam Score')
plt.show()
```

Correlation Heatmap

```python
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='Blues')
plt.title('Correlation Heatmap')
plt.show()
```

As we can see Hours_studied and attendance have more correlation with exam score.

so we can say that hours and attendace have much impact the scores of exam
